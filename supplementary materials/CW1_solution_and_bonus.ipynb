# A. Поездки на метро

__Почему оказалась в контрольной:__ простая задачка для разминки. На курсере вам просили решить много подобных задач с придумыванием формул. Эта задача была по их мотивам. 

------


На некоторых кросс-платформенных станциях метро (как, например, "Третьяковская") на разные стороны платформы приходят поезда разных направлений. Таня договорилась встретиться с подругой на такой станции, но поскольку подруга приехала из другого часового пояса, то из-за джетлага сильно проспала, и Тане пришлось долго её ждать. Поезда всегда ходят точно по расписанию, и Таня знает, что поезд стоит на платформе ровно одну минуту, а интервал между поездами (время, в течение которого поезда у платформы нет) составляет $a$ минут для поездов на первом пути и $b$ минут для поездов на втором пути. То есть на первый путь приезжает поезд и стоит одну минуту, затем в течение $a$ минут поезда у платформы нет, затем в течение одной минуты у платформы стоит следующий поезд и т.д.

Пока Таня стояла на платформе, она насчитала $n$ поездов на первом пути и $m$ поездов на втором пути. Определите минимальное и максимальное время, которое Таня могла провести на платформе, или сообщите, что она точно сбилась со счёта.  

Все поезда, которые видела Таня, она наблюдала в течение всей минуты, то есть Таня не приходит и не уходит с платформы посередине той минуты, когда поезд стоит на платформе.

```python
a = int(input())
b = int(input())
n = int(input())
m = int(input())

l = max(n + (n - 1) * a, m + (m - 1) * b)
r = min(n + (n + 1) * a, m + (m + 1) * b)

if l <= r:
    print(l, r)
else:
    print(-1)
```

# Б. Ряд Тэйлора для arctan

__Почему оказалась в контрольной:__ простая задача на цикл.

-----

Никто не знает зачем, но Винни-Пух хочет используя ряд Тейлора вычислить значение функции $\arctan x$ в данной точке.

$$
\arctan x \approx x - \frac{x^3}{3} + \ldots + \frac{(-1)^n x^{2n+1}}{2n + 1} + \ldots
$$

Сова строго-настрого запретила ему использовать библиотеку math. Помогите Винни-Пуху. 

```python
x = float(input())

res = 0
for n in range(500):
    res += (-1)**n * x**(2 * n + 1) / (2 * n + 1)
    
print(res)
```

# C. Горшки для хвоста

__Почему оказалась в контрольной:__ один в один такая же задача была на курсере. Она называлась "Обувной магазин". Кто решал курсеру самостоятельно получал здесь приемущество и делал эту задачу легко и быстро.

-----

Винни-Пух собирается подарить Иа-Иа в подарок коллекцию горшочков. Тогда Иа-Иа сможет положить в него свой хвост.  

У Совы есть много горшочков разного размера. Винни-Пух думает, что если положить одни горшок внутрь другого, а потом уже хвост, то он будет лучше защищён. Известно, что один горшок можно положить внутрь другого, если он хотя бы на три размера больше. 

Какое наибольшее количество горшков Сова может предложить Винни-Пуху, чтобы он смог положить их друг в друга и защитить хвост Иа-Иа? 
   

```python
b = int(input())
items = list(map(int, input().split()))

items.sort() # чтобы все горшки шли по порядку
c = b - 3
d = 0

for item in items:
    if item - c >= 3:
        c = item
        d += 1
        
print(d)
```

------

Коронавирус повсюду! Он даже проник в контрольную. И теперь живёт в клетках квадрата размера $N$ x $N$ (не больше одного в клетке). Каждую минуту обстанова внутри контрольной меняется и происходит следующее: 

- все коронавирусы, у которых более 3-х соседей, умирают от перенаселения (соседями будем называть вирусы, которые живут в клетках, имеющих общую сторону или вершину);

- все коронавирусы, у которых меньше 2-х соседей, умирают от скуки;

- в клетках, у которых ровно 2 соседа ничего не происходит;

- на всех пустых клетках, у которых ровно в трёх соседних клетках жили коронавирусы, появляются новые коронавирусы.

Все изменения происходят одновременно. То есть сначала для каждой клетки решается что именно с ней произойдёт в следующую минуту, а затем изменения вступают в силу. 

Требуется по заданной конфигурации понять, во что она превратится через $T$ минут.

```python
def get_neis(x, y):
    cnt = - field[y][x]
    for i in range(-1, 2):
        for j in range(-1, 2):
            cnt += field[y + i][x + j]
    return cnt


n, t = list(map(int, input().split()))
inf = 1000

field = [list(map(int, input().split())) + [0] for i in range(n)]
field = field + [[0 for i in range(n + 1)]]
new_field = [field[i][::] for i in range(n + 1)]

for i in range(t):
    for y in range(n):
        for x in range(n):
            neis = get_neis(x, y)
            
            if (neis < 2 or neis > 3):
                new_field[y][x] = 0
            
            if (neis == 3):
                new_field[y][x] = 1
    
    field = [new_field[i][::] for i in range(n + 1)]

for i in field[:-1]:
    print(*i[:-1])
```

# E. Дорожные Винни-Пухи

__Почему оказалась в контрольной:__ Дробящая мозг задача, над которой надо было подумать. Была расчитана на тех, кто собрался получать дополнительные баллы.

------

Вдоль дороги идут три Винни-Пуха. 

- Один Винни-Пух говорит: “Впереди меня два Винни-Пуха”. 

- Другой Винни-Пух говорит:  “Позади меня два Винни-Пуха”.

- Третий Винни-Пух говорит: “Впереди меня два Винни-Пуха и позади меня два Винни-Пуха”.

Как такое может быть? Ответ: третий Винни-Пух врёт! 

По дороге один за другим движутся N Винни-Пухов. Каждый из них говорит фразу вида: “Впереди меня $a_i$ Винни-Пухов, а позади меня $b_i$ Винни-Пухов”.

Ваша задача определить какое самое большее количество Винни-Пухов могут говорить правду.


```python
n = int(input())

r = 0
mat = [1]*n

for i in range(n):
    st = input()
    front, back = (int(a) for a in st.split())
    if min(front, back) >= 0 and front + back == n - 1:
        if front < n and mat[front] != 2:
            mat[front] = 2
            r += 1
print(r)
```

Если вы ходили на семинары и на консультации к асистентам, проблем возникнуть было не должно, так как семинарский код адаптировался под контрольную дописыванием лишнего цикла. Решение в лоб через три цикла принималось за $12$ баллов из $30$. 

-------

У Кролика есть три множества горшочков с мёдом: $A$, $B$ и $C$. Все горшочки содержат разное количество мёда. Винни-Пух пришёл в гости и хочет съесть $S$ килограмм мёда.

Кролику нужно найти все такие тройки чисел $(a, b, c)$ из трёх разных множеств, которые в сумме дадут $S$: $a + b + c = S$.

```python
def slow_algo(A,B,C,S):
    q = 0
    for i in range(len(A)):
        for j in range(len(B)):
            for k in range(len(C)):
                if A[i] + B[j] + C[k] == S:
                    q += 1
    return q
```

```python
def fast_algo(A,B,C,S):
    q = 0
    C = set(C)
    for i in range(len(A)):
        for j in range(len(B)):
            if S - A[i] - B[j] in C:
                q += 1
    return q
```

```python
n_a = int(input())
A = list(map(int, input().split(" ")))

n_b = int(input())
B = list(map(int, input().split(" ")))

n_c = int(input())
C = list(map(int, input().split(" ")))

S = int(input())

print(fast_algo(A,B,C,S))
```

Винни-Пух услышал, что прогресс не стоит на месте, продал пяточка на чёрном рынке и купил себе несколько видеокарт для обучения нейронных сетей. 

Винни-Пух учит $n$ нейросеток. По одной на каждой видеокарте. Все видеокарты работают независимо друг от друга. Время обучения на каждой из них равномерно распределено на отрезке $[0; 3]$ минут. В процессе обучения видеокарта может сгореть. Время $T$ через которое сгорает любая видеокарта имеет экспоненциальное распределение с параметром $\lambda = 0.3$. 

Все карты горят независимо друг от друга.  Время, через которое видеокарта сгорает не зависит от времени обучения сетки. Известно, что сгорела ровно одна видеокарта. Оцените с помощью симуляций вероятность того, что $T \le 1$. 

```python
import random
```

```python
import random as rd
n = 10
one = 0

both=0
p = 3*10**5

for i in range(p):

    count = 0
    for ii in range(n):
        learn = rd.uniform(a=0, b=3)
        gor = rd.expovariate(lambd=0.3)
        if learn > gor:
            count +=1
            gor2 = gor
    if count == 1:
        one += 1
    if gor2 <= 1 and count == 1:
            both += 1

print(both/one)
```

То же самое через кучу листов. 

```python
n = 10
n_obs = 100000

# первый массив сколько работала сетка, второй когда потенциально могла упасть 
NN = [[random.uniform(0,3) for i in range(n)] for i in range(n_obs)] 
T = [[random.expovariate(0.3) for i in range(n)] for i in range(n_obs)]

# сколько карт сгорело в какой серии испытаний
fire = [sum([item > t for t,item in zip(tt, nn)]) for tt,nn in zip(T, NN)]

# подходит ли нам испытание
nice_T = [sample for i,sample in enumerate(T) if fire[i] == 1]
nice_N = [sample for i,sample in enumerate(NN) if fire[i] == 1]

# время до сгорания в хороших испытаниях
sample = [t for tt,nn in zip(nice_T, nice_N) for t,n in zip(tt,nn) if t < n]

sum([sm < 1 for sm in sample])/len(sample), len(sample)
```

То же самое, но через `numpy`. На которнольной так было делать нельзя. Но тем, кто будет смотреть, рекомендую разобраться. Через numpy выглядит попроще. 

```python
import numpy as np

n = 10
n_obs = 100000

NN = np.random.uniform(0,3, (n_obs, n))
T = np.random.exponential(1/0.3, (n_obs, n))  # пояснение про параметр ниже!

fire = np.sum(NN > T, axis=1)

nice_N = NN[fire == 1]
nice_T = T[fire == 1]

# проверка размерностей для себя
assert nice_T.shape == nice_N.shape

sample = nice_T[nice_T < nice_N]

assert sample.shape[0] == nice_T.shape[0]

np.mean(sample < 1)
```
